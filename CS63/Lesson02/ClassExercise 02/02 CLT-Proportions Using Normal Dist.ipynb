```python
from datascience import *
import numpy as np
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
%matplotlib inline

import random
```

```python
# Suppose candidate 'A' has 60% support in the population

population_proportions = make_array('A', 'A', 'A', 'A', 'A', 'A', 'B', 'B','B','B')

model = Table().with_column('Vote', population_proportions)

model
```

```python
# Take a sample of voters and measure the proportions

num_voters = 10
sample = model.sample(num_voters)
print('Sample of 10 voters =',sample)

A_count = np.count_nonzero(sample.column('Vote') == 'A')
print('Total Voters supporting A =',A_count)

prop = A_count/num_voters
print('Proportion = ',prop)
```

```python
props = make_array()

num_voters = 500  # Sample size
repetitions = 10000

for i in np.arange(repetitions):
    sample = model.sample(num_voters)
    new_prop = np.count_nonzero(sample.column('Vote') == 'A')/num_voters
    props = np.append(props, new_prop)
    

results = Table().with_column('Sample Voter Proportions', props)
print(results)
results.hist(bins=np.arange(0.3, 0.85, 0.01))


print('Standard Deviation of Proportions =',props.std())


# Suppose candidate 'A' has 60% support in the population
p = 0.6
q = 1-p
print("Theoretical Value = ", np.sqrt(p*q/num_voters))
```
