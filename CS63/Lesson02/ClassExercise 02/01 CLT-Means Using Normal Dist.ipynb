```python
from datascience import *
import numpy as np
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
%matplotlib inline

import random
```

```python
# Generate 100,000 random numbers from normal distribution with mu=0, sigma=3
#
numbers_to_generate = 100000

mean_of_population = 0
std_of_population = 3


num_array = make_array()

for i in range(numbers_to_generate):
    num = random.normalvariate(mean_of_population,std_of_population)
    num_array = np.append(num_array, num)

population = Table().with_column('Data', num_array)
print(population)

population.hist(bins=100)
population.hist(bins=np.arange(-3*std_of_population, 3*std_of_population, 0.25))
```

```python
sample_size = 500
number_of_samples = 1000

means_of_samples = make_array()

for i in np.arange(number_of_samples):
    sample = population.sample(sample_size)
    new_mean = np.mean(sample.column('Data'))
    means_of_samples = np.append(means_of_samples,new_mean)

results = Table().with_column(
    'Sample Mean', means_of_samples
)
results.hist(bins=50)

print("Standard Deviation of Means =",means_of_samples.std())

print("Theoretical Value = ", std_of_population/np.sqrt(sample_size))
```
